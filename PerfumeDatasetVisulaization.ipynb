## Data loading

### Subtask:
Load the data from "Indonesian Local Perfume Brands.csv" into a pandas DataFrame.


**Reasoning**:
Load the data from the provided CSV file into a pandas DataFrame and display its first few rows and shape.



```python
import pandas as pd

try:
    df = pd.read_csv('Indonesian Local Perfume Brands.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'Indonesian Local Perfume Brands.csv' not found.")
    df = None
except Exception as e:
    print(f"An error occurred: {e}")
    df = None
```

## Data exploration

### Subtask:
Explore the dataset to identify a suitable categorical or continuous variable for visualization and determine its distribution.


**Reasoning**:
I need to explore the dataset to identify a suitable categorical or continuous variable for visualization and determine its distribution. This involves examining data types, checking for missing values and unique values, calculating descriptive statistics, and choosing an appropriate variable for visualization.



```python
# Examining data types and identify potential variables
print(df.dtypes)

# Checked for missing values and unique values
for col in ['brand', 'followers', 'price', 'concentration', 'top_notes1', 'mid_notes1', 'base_notes1']:
    print(f"\nColumn: {col}")
    print(f"Missing values: {df[col].isnull().sum()}")
    print(f"Unique values: {df[col].nunique()}")
    print(f"Unique values: {df[col].unique()[:10]}")

# Calculated descriptive statistics for continuous variables
print("\nDescriptive statistics for 'followers' and 'price':")
# Convert 'price' to numeric, handling errors
df['price'] = pd.to_numeric(df['price'].str.replace(r'[.,]', '', regex=True), errors='coerce')
print(df[['followers', 'price']].describe())

# Calculated frequencies for categorical variables
print("\nFrequency counts for 'brand' and 'concentration':")
print(df['brand'].value_counts().head(10))
print(df['concentration'].value_counts())


# Choosing a variable for visualization
# Justification: 'brand' is a categorical variable with a reasonable number of unique values and no missing data.
# It's informative to visualize the distribution of brands in the dataset.
chosen_variable = 'brand' #Example, can be changed based on the analysis
print(f"\nChosen variable for visualization: {chosen_variable}")
```

## Data preparation

### Subtask:
Prepare the 'brand' column for visualization.


**Reasoning**:
Check for missing values in the 'brand' column, convert its data type to categorical if necessary, and then verify the data is ready for visualization.



```python
# Checked for missing values in the 'brand' column
missing_before = df['brand'].isnull().sum()
print(f"Number of missing values in 'brand' before handling: {missing_before}")

# No missing values found in the previous exploration step, so skipping imputation or removal.

# Checked the data type of the 'brand' column
print(f"Data type of 'brand' before conversion: {df['brand'].dtype}")

# Converted to categorical if necessary
if df['brand'].dtype != 'object':
    df['brand'] = df['brand'].astype('category')

print(f"Data type of 'brand' after conversion: {df['brand'].dtype}")

# Verified the data is ready for visualization
print("\nFirst 10 unique values in 'brand':")
print(df['brand'].unique()[:10])
print(f"\nTotal number of unique values in 'brand': {df['brand'].nunique()}")
```

## Data visualization

### Subtask:
Visualizing the distribution of perfume brands using a bar chart.


**Reasoning**:
Visualized the distribution of perfume brands using a bar chart with appropriate labels and title, and rotate x-axis labels for better readability if necessary. Adjust the figure size to accommodate all brand names clearly.



```python
import matplotlib.pyplot as plt

# Calculated brand frequencies
brand_counts = df['brand'].value_counts()

# Created the bar chart
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(brand_counts.index, brand_counts.values, color='skyblue')
plt.xlabel("Brand Name", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Indonesian Local Perfume Brands", fontsize=14)
plt.xticks(rotation=90, ha='right', fontsize=8)  # Rotate x-axis labels
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```
